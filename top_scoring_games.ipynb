```python
import json
import pandas as pd
```

```python
league = 48
```

```python
with open(f'data/{league}/league_details.json','r') as f:
    data = json.loads(f.read())
```

```python
teams = {}

for team in data['league_entries']:
    id = team['id']
    name = team['entry_name']

    if not id in teams:
        teams[id] = name
```

```python
results = []

for game in data['matches']:
    week = game['event']
    entry1 = game['league_entry_1']
    entry1points = game['league_entry_1_points']
    entry2 = game['league_entry_2']
    entry2points = game['league_entry_2_points']

    team1 = teams[entry1]
    team2 = teams[entry2]

    results.append([team1, entry1points, team2])
    results.append([team2, entry2points, team1])
```

```python
df = pd.DataFrame(results, columns=['team', 'points', 'against'])
df = df.sort_values(by='points', ascending=False)
df.groupby('team').head(5).sort_values(by=['team', 'points'], ascending=[True, False])
```
